```python
import os, numpy as np, pandas as pd, matplotlib.pyplot as plt

%matplotlib inline

data = pd.read_csv("college-major-data/recent-grads.csv")

recent_grads = data.dropna()
```

# Question:
Do students in more popular majors make more money?

```python
ax1 = recent_grads.plot(x = 'Median', y = 'Sample_size',
                       kind = 'scatter', title = 'Sample size vs. Median')
ax1.set_ylabel('Sample size')
```

# Conclusion:
A smaller available sample would suggest lower popularity in the chosen major. If popularity of the major was the only factor that had an impact on a student's salary we would expect there to be some correlation between the sample size and median salary. It seems to be the case that the least popular majors offer both the highest and lowest median salaries, this is likely due to the fact that some unpopular majors are highly challenging and offer a much sought after set of skills; or in contrast, don't teach students skills that are highly valued by employers.

# Question:
Do students that majored in subjects that were majority female make more money?

```python
ax2 = recent_grads.plot(x = 'Median', y = 'ShareWomen',
                       kind = 'scatter', title = '% Women vs. Median')
ax2.set_ylabel('% Women')
```

# Conclusion:
Plotting the % of women taking each major against median salary we can clearly see a negative correlation. The majors who had a greater proportion of female students offered lower median salaries.

# Question:
Is there any link between the number of full-time employees and median salary?

```python
ax3 = recent_grads.plot(x = 'Median', y = 'Full_time',
                       kind = 'scatter', title = 'Full time vs. Median')
ax3.set_ylabel('Full time')
```

# Conclusion:
There is a weak negative correlation for a subset of majors with a high full time employment, however, majority of the majors seem to cluster at the lower end of the salary scale and provide fewer students with full time employment.

# Question:
What percent of majors are predominantly male? Predominantly female?

```python
ax4 = recent_grads['ShareWomen'].plot(kind = 'hist', bins = 10)
ax4.set_title('Distribution of women as proportion/major')
```

# Conclusion:
There seems to be an approximately 50/50 split between majors that are predominantly taken by females and majors that are predominantly taken by males.

# Question:
What's the most common median salary range?

```python
ax5 = recent_grads['Median'].plot(kind = 'hist', rot = 45, bins = 10)
ax5.set_title('Distribution of median salary/major')
```

# Conclusion:
Clearly the vast majority of majors offer median salaries ranging between ~ \$30,000 - \$40,000.

# Question
Compare the percentages of women and unemployment rates from the 10 highest paying majors and from the 10 lowest paying majors.

```python
fig, axes = plt.subplots(figsize = (10,10), nrows = 2, ncols = 2)

recent_grads[:10].plot(x = 'Major', y = 'ShareWomen', legend = False, ax = axes[0,0], kind = 'bar')
recent_grads[:10].plot(x = 'Major', y = 'Unemployment_rate', legend = False, ax = axes[1,0], kind = 'bar')

recent_grads_sorted = recent_grads.sort_values('Median', ascending = True) # sort dataframe by ascending median salary

recent_grads_sorted[:10].plot(x = 'Major', y = 'ShareWomen', legend = False, ax = axes[0,1], kind = 'bar')
recent_grads_sorted[:10].plot(x = 'Major', y = 'Unemployment_rate', legend = False, ax = axes[1,1], kind = 'bar')

for i in range(0,2):
    axes[0,i].tick_params(labelbottom='off')
    axes[0,i].set_xlabel('')
    axes[0,i].set_ylabel('% Women')
    axes[0,i].set_ylim(0,1)
    axes[1,i].set_ylim(0,0.18)
    axes[1,i].set_ylabel('Unemployment rate')

axes[1,0].set_xlabel('Highest paying majors')
axes[1,1].set_xlabel('Lowest paying majors')
    
    
```

# Conclusion
From the charts we can clearly see that the ten lowest paying majors are predominantly chosen by women while the ten highest paying majors are predominantly chosen by men with the exception of actuarial science.

On average unemployment rates seem to be slightly lower for the ten highest paying majors.
